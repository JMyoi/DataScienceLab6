Exercise 1: Line Plot for Total Profit Over Months


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('company_sales_data.csv')

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['total_profit'], marker='o')
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

Exercise 2: Styled Line Plot for Total Profit


```python
plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['total_profit'], linestyle='--', color='red', marker='o', markerfacecolor='red', linewidth=3, label='Profit data')
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

Exercise 3: Multi-line Plot for Product Sales


```python
plt.figure(figsize=(10, 5))
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in products:
    plt.plot(df['month_number'], df[product], marker='o', label=product)
plt.title('Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(True)
plt.show()
```

Exercise 4: Scatter Plot for Toothpaste Sales


```python
plt.figure(figsize=(10, 5))
plt.scatter(df['month_number'], df['toothpaste'], color='blue')
plt.title('Toothpaste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.grid(True, linestyle='--')
plt.show()
```

Exercise 5: Bar Chart for Face Cream and Facewash Sales


```python
plt.figure(figsize=(10, 5))
width = 0.35  # the width of the bars
p1 = plt.bar(df['month_number'] - width/2, df['facecream'], width, label='Face Cream')
p2 = plt.bar(df['month_number'] + width/2, df['facewash'], width, label='Face Wash')

plt.title('Face Cream and Face Wash Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```

Exercise 6: Bar Chart for Bathing Soap Sales


```python
plt.figure(figsize=(10, 5))
plt.bar(df['month_number'], df['bathingsoap'], color='blue')
plt.title('Monthly Sales of Bathing Soap')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.savefig('bathing_soap_sales.png')  # Saving the plot
plt.show()
```

Exercise 7: Histogram of Total Profit


```python
plt.figure(figsize=(10, 5))
plt.hist(df['total_profit'], bins=10, color='green', alpha=0.7)
plt.title('Histogram of Total Profit')
plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.show()
```

Exercise 8: Pie Chart for Total Annual Sales by Product


```python
total_sales = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
labels = total_sales.index
sizes = total_sales.values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Annual Sales Distribution by Product')
plt.show()
```

Exercise 9: Subplot for Bathing Soap and Facewash Sales


```python
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

ax[0].plot(df['month_number'], df['bathingsoap'], marker='o', color='blue')
ax[0].set_title('Monthly Sales of Bathing Soap')
ax[0].set_xlabel('Month Number')
ax[0].set_ylabel('Units Sold')

ax[1].plot(df['month_number'], df['facewash'], marker='o', color='green')
ax[1].set_title('Monthly Sales of Facewash')
ax[1].set_xlabel('Month Number')
ax[1].set_ylabel('Units Sold')

plt.tight_layout()
plt.show()
```

Exercise 10: Stack Plot for All Products


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('company_sales_data.csv')

# Prepare data for plotting
plt.figure(figsize=(10, 5))
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
data = [df[product] for product in products]

# Create stack plot
plt.stackplot(df['month_number'], *data, labels=products)
plt.title('Total Product Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')
plt.show()
```
